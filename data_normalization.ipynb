```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import glob
import cv2
from sklearn.model_selection import train_test_split
```

```python
class_names = ["EMCI","LMCI","MCI","AD","CN"]
```

```python
path=r"G:\dataset\Alzheimers-ADNI"
```

```python
trainData=[]
trainDataLables=[]
for name in class_names:
    images = [cv2.imread(file,cv2.COLOR_RGB2GRAY) for file in glob.glob(path+"\\train\\"+name+r'\*.jpg')]
    for i in images:
        i=cv2.resize(i,(256, 256))
        trainData.append(np.asarray(i))
        trainDataLables.append(name)


trainData=np.asarray(trainData)
trainDataLables=np.asarray(trainDataLables)
```

```python
now=trainData[0]

plt.imshow(now)
```

```python
hist = cv2.calcHist([now], [0], None, [256], [0, 256])

plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.plot(hist)
plt.xlim([0, 256])
```
